```python
import pandas as pd
#series = a pandas 1D labeled array capable of holding any data type
#        think of it as a single column in a spreadsheet(1D)

data = [100.1, 200.2, 300.3, 400.3, 500.4,"abc", True]
series = pd.Series(data) #it has capital S
print(series)
```

```python
data = [100, 200, 300, 400, 500]
series = pd.Series(data, index=['a', 'b', 'c', 'd', 'e']) #custom index
print(series)
print(series.loc['c']) #accessing an element by index label
series.loc["a"]= 111 #modifying an element by index label
print(series)
print(series.iloc[2]) #accessing an element by index position
print(series[series > 300]) #filtering
```

```python
import pandas as pd
calories={"day1":420,"day2":380,"day3":390}
series=pd.Series(calories)
print(series)
print(series['day1'])
series.loc['day1']+=1000
print(series)
print(series[series>400])
```
