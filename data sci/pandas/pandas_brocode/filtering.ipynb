```python
import pandas as pd
df=pd.read_csv('pokemon.csv')
#FILTERING =keeping rows that meet criteria
tall_pokemon=df[df['Height']>=2]
print(tall_pokemon.to_string())
```

```python
import pandas as pd
df=pd.read_csv('pokemon.csv')
heavy_pokemon=df[df['Weight']>=100]
print(heavy_pokemon.to_string())
```

```python
import pandas as pd
df=pd.read_csv('pokemon.csv')
legendary_pokemon=df[df['Legendary']==True]
print(legendary_pokemon.to_string())
```

```python
import pandas as pd
df=pd.read_csv('pokemon.csv')

water_pokemon=df[(df['Type1']=='Water')|(df['Type2']=='Water')]
print(water_pokemon.to_string())
```

```python
import pandas as pd
df=pd.read_csv('pokemon.csv')
# fire and flying type pokemon
fire_flying_pokemon=df[(df['Type1']=='Fire')&(df['Type2']=='Flying')]
print(fire_flying_pokemon.to_string())
```

```python
import pandas as pd
df=pd.read_csv('pokemon.csv')
#dragon and legendary pokemon
dragon_legendary_pokemon=df[(df['Type1']=='Dragon')&(df['Legendary']==True)]
print(dragon_legendary_pokemon.to_string())
```
