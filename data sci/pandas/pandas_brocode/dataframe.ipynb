```python
import pandas as pd
data={"name":["songebob","patrick","sandy"],
      "age":[30,35,50]
      }
df=pd.DataFrame(data,index=['emp #1','emp #2','emp #3'])
print(df)
print(df.loc['emp #2'])
#add a new column
df['job'] = ["cook", "n/a", "customizer"]
print(df)
#add a new row
new_row = pd.DataFrame({"name": "squidward", "age": 40, "job": "lifeguard"}, index=['emp #4'])
df = pd.concat([df, new_row])
print(df)
#add new rows
new_rows = pd.DataFrame([{"name": "squidward", "age": 40, "job": "lifeguard"},
                        {"name": "shri", "age": 21, "job": "batman"}], index=['emp #4', 'emp #5'])
df = pd.concat([df, new_rows])
print(df)
```
