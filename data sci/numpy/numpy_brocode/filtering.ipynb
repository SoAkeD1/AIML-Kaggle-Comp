```python
#filtering = refering to the process of selecting elements
#            from an array that match a given condtion
```

```python
import numpy as np
ages=np.array([[21,17,19,20,16,30,18,65],
               [39,22,15,99,18,19,20,21]])
teenagers=ages[ages<18]
adults=ages[(ages>=18) & (ages<65)]
print(adults)
print(teenagers)
adults=np.where(ages>=18, ages , 0)
print(adults)
```
