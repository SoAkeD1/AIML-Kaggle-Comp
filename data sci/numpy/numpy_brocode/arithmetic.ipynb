```python
import numpy as np

#scalar arithmetric
array=np.array([1,2,3])
print(array-1)
print(array*3)
print(array**5)
print(array/3)
```

```python
import numpy as np
#vectorized math function
array=np.array([1.01,2.34,3.09])
print(np.sqrt(array))
print(np.round(array))
print(np.pi)
print(np.floor(array))
print(np.ceil(array))


```

```python
import numpy as np
#vectorized math function
radii=np.array([1,2,3])
print(np.pi*radii**2)
```

```python
import numpy as np
#elementwise artithmetric
array1=np.array([1,2,3])
array2=np.array([4,5,6])
print(array1+array2)
print(array1/array2)
print(array1*array2)
print(array1**array2)
```

```python
import numpy as np
#comparsion operator
scores=np.array([91,55,100,73,82,64])
print(scores==100)
print(scores>60)
scores[scores<60]=0
print(scores)
```
