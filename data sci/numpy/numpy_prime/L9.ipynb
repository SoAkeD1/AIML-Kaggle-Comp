```python
import numpy as np  
arr=np.array([1,2,3.14,4,5])
print(arr, arr.dtype)
#system dependent int32 or int64 , in mac it is int32 and windows it is int64
```

```python
#image  reading --->black and white images ---> 0 to 255
#black =0 , 255 =white , 1 to 224 = different shades of gray
#complex no. in numpy 2 parts - real and imaginary , eg: 3+4j
import numpy as np
arr1=np.array([1,2,3,4], dtype="complex64")
print(arr1, arr1.dtype)
```

```python
arr=np.array([3+4j])
arr1=np.array([3+7j])
print(arr, arr.dtype)
print(arr+arr1)
```
