```python
#normalization
#means: normalizing an array means transforming its values to a common scale
# , without distorting differences in the ranges of values.
import numpy as np
arr=np.array([[1,2],[3,4]])
print(np.mean(arr))  #2.5
std_dev=np.std(arr)   #1.118033988749895
normalized_arr=(arr-np.mean(arr))/std_dev
print(normalized_arr)
#mean=0 and std=1 , here

```
