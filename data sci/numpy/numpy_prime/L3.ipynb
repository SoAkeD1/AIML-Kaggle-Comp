```python
import numpy as np
import time
size=10_000_000
#python list
py_list=list(range(size))
start=time.time()
sq_list=[x*2 for x in py_list]
end=time.time()
print("Python list time:",end-start)

# np array

np_arr=np.array(py_list)
start=time.time()
sq_arr=np_arr*2
end=time.time()
print("Numpy array time:",end-start)
```

```python
import sys
print("Python list size (bytes):",sys.getsizeof(py_list)*len(py_list))
print("Numpy array size (bytes):",np_arr.nbytes)
```
