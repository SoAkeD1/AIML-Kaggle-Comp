```python
import numpy as np  
arr=np.array([1,2,3,4,5])
#[start:stop:step]
print(arr[1:4:2])  # Output: [2 4]
print(arr[::2])    # Output: [1 3 5]
print(arr[1::2])   # Output: [2 4]  
```
