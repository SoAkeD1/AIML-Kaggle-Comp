```python
#aggregation functions in numpy
import numpy as np
arr=np.array([[1,2,3,4,5]])
print(np.sum(arr))  #sum of all elements
print(np.prod(arr)) #product of all elements
print(np.mean(arr)) #mean of all elements
print(np.max(arr))  #maximum element
print(np.min(arr))  #minimum element
print(np.argmin(arr))  #index of minimum element
print(np.argmax(arr))  #index of maximum element
print(np.std(arr))     #standard deviation
print(np.median(arr))  #median
print(np.percentile(arr, 50)) #50th percentile
```

```python
#POWER FUNCTION
import numpy as np
arr=np.array([1,2,3,4,5])
print(np.power(arr, 3))  #cube of each element
print(np.sqrt(arr))      #square root of each element
print(np.cbrt(arr))      #cube root of each element
print(np.log(arr))       #natural logarithm of each element
print(np.log10(arr))     #base-10 logarithm of each element
print(np.log2(arr))      #base-2 logarithm of each element
print(np.exp(arr))       #exponential of each element
```

```python
#ROUNDING FUNCTIONS
import numpy as np
arr=np.array([1.2, 2.5, -3.7, 4.4, 5.9])
print(np.around(arr))   #round to nearest integer
print(np.floor(arr))    #round down to nearest integer
print(np.ceil(arr))     #round up to nearest integer
print(np.trunc(arr))    #truncate decimal part
print(np.rint(arr))     #round to nearest integer, .5 rounds to even
print(np.fix(arr))      #round towards zero
print(np.modf(arr))     #separate fractional and integer parts
print(np.unique(arr))  #find unique elements
print(np.sort(arr))    #sort the array
print(np.abs(arr))     #absolute values

```
