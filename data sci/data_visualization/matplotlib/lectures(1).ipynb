```python
import matplotlib.pyplot as plt
# types of scatter plots : 1. corelated ,2. outlies , 3. clusters
people = ["Person A", "Person B", "Person C", "Person D", "Person E", "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]


colors = ["green" if x<135 else "red" for x in bp]
plt.scatter(age,bp, color=colors)
plt.title("age VS bp")
plt.xlabel("age")
plt.ylabel("blood-pressure")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
#color bar
plt.colorbar(label="BP")
```

```python
#annotations
#plt.annotate("text",xy=(x_point,y_point),xytext=(x_text,y_text)
import matplotlib.pyplot as plt

people = ["Person A", "Person B", "Person C", "Person D", "Person E", "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]


colors = ["green" if x<135 else "red" for x in bp]
plt.scatter(age,bp, color=colors)
plt.title("age VS bp")
plt.xlabel("age")
plt.ylabel("blood-pressure")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
#color bar
plt.colorbar(label="BP")
for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext=(age[i]+1,bp[i]+1))
plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+20)
```
