```python
#decision tree regressor
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
```

```python
df=load_diabetes(as_frame=True).frame
df.head()
#this is a preprocessed dataset, so we can directly apply the model on it. We don't have to do any data cleaning or feature engineering. We can directly apply the model on it.
```

```python
X=df.drop('target',axis=1)
y =df['target']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=3, random_state=42,  min_samples_leaf=10)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred_train=model.predict(X_train)
y_pred_test=model.predict(X_test)

print("MSE train:", mean_squared_error(y_train, y_pred_train))
print("MSE test:", mean_squared_error(y_test, y_pred_test))
print("r2 train:", r2_score(y_train, y_pred_train))
print("r2 test:", r2_score(y_test, y_pred_test))
```

```python
from sklearn.tree import plot_tree
plot_tree(model, feature_names=X.columns, filled=True, rounded=True)
plt.figure(figsize=(18,10))
plt.tight_layout
```
